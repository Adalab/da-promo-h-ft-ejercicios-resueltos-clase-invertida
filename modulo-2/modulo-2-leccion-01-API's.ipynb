**Sacamos información de los equipos**

```python
import requests
import pandas as pd

# Inicializar el diccionario de equipos
equipos = {"conference": [], "name": [], "city": [], "abbreviation": []}

# Realizar la solicitud HTTP a la API de equipos
res_equipos = requests.get(f"http://api.balldontlie.io/v1/teams", headers = {'Authorization': api_key}) 

# Verificar si la solicitud fue exitosa
if res_equipos.status_code == 200:
    # Convertir la respuesta JSON en un diccionario de Python
    res_equipos = res_equipos.json()
    
    # Iterar sobre los equipos y agregar información al diccionario de equipos
    for equipo in res_equipos["data"]:
        equipos["conference"].append(equipo["conference"])
        equipos["name"].append(equipo["full_name"])
        equipos["city"].append(equipo["city"])
        equipos["abbreviation"].append(equipo["abbreviation"])
    
    # Convertir el diccionario de equipos en un DataFrame de Pandas
    df_equipos = pd.DataFrame(equipos)
    
    # En lugar de imprimir el DataFrame, simplemente guárdalo en una variable
else:
    print("Error al obtener los datos de los equipos.")

df_equipos
```
