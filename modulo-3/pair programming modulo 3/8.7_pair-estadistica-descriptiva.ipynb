```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("world_data_full_apply_nulos.csv", index_col = 0)
df.head()
```

```python
# 1. Calcula las siguientes estadísticas descriptivas para las variables numéricas:
def estadisticas (dataframe, col): 
    media = dataframe[col].mean()
    mediana = dataframe[col].median()
    desviacion  = dataframe[col].std()
    minimo = dataframe[col].min()
    maximo = dataframe[col].max()
    
    print(f"La media de la columna {col} es {media}")
    print(f"La mediana de la columna {col} es {mediana}")
    print(f"La desviación de la columna {col} es {desviacion}")
    print(f"El minimo de la columna {col} es {minimo}")
    print(f"La maximo de la columna {col} es {maximo}")
    print("...................")
```

```python
col_num = df.select_dtypes(include = np.number)
for columna in col_num:
    estadisticas(df, columna)
```

```python
#  2. Identifica el país con la mayor población y el país con la menor población en el conjunto de datos
max_poblacion = df[df["density"] == df["density"].max()]["country"].values[0]
min_poblacion = df[df["density"] == df["density"].min()]["country"].values[0]

print(f"La poblacion con mayor densidad es {max_poblacion} y el que menos es {min_poblacion}")
```

```python
# 3. Calcula la tasa promedio de natalidad (birth_rate) para todos los países en el conjunto de datos.
df.groupby("country")["birth_rate"].mean().reset_index()
```

```python
# 5. Calcula la correlación entre las variables de "fertility_rate" y "life_expectancy". Interpreta la relación entre estas dos variables.
df[["fertility_rate", "life_expectancy"]].corr()
```

```python
# 6. Visualiza la distribución de la variable "cpi" utilizando un histograma y comenta sobre la distribución de los índices de percepción de corrupción.
sns.histplot(x = "cpi", data = df)
plt.xlabel("Índice de Percepción de Corrupción (CPI)")
plt.ylabel("Frecuencia")
plt.title("Distribución del Índice de Percepción de Corrupción");
```

```python
# 7. Encuentra el país con la mayor diferencia entre el "gdp" (Producto Interno Bruto) y el "co2-emissions" (emisiones de dióxido de carbono).
df["diferencia_gdp_co2"] = df["gdp"] - df["co2-emissions"]
df[df["diferencia_gdp_co2"] == df["diferencia_gdp_co2"].max()]["country"]
```

```python
# 8. Calcula el porcentaje promedio de mujeres que participan en la fuerza laboral (population_labor_force_participation) para todos los países.
# Calcula el porcentaje promedio de mujeres en la fuerza laboral
porcentaje_promedio_mujeres = df["population_labor_force_participation"].mean()

# Imprime el resultado
print(f"El porcentaje promedio de mujeres en la fuerza laboral es: {porcentaje_promedio_mujeres:.2f}%")
```

```python
# 9. Genera un mapa de calor (heatmap) de la matriz de correlación de todas las variables numéricas y discute las relaciones más destacadas.sns.

plt.figure(figsize = (15, 15))

# creamos una matriz booleana de máscara
mask = np.triu(np.ones_like(df.corr(numeric_only=True), dtype=bool))

# creamos un heatmap con la máscara
sns.heatmap(df.corr(numeric_only=True), 
            annot = True, 
            fmt = ".2f", 
            cmap = "mako", 
            vmax = 1, 
            vmin = -1, 
            mask = mask); 
```
