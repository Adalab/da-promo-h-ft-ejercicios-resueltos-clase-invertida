# *Pair programming* clases

```python
class Empleados:
    def __init__(self, nombre, apellidos, edad, posicion, año, vacaciones, herramientas):
        
        # estamos creando un atributo llamado nombre y asignandoselo al valor del parámetro nombre
        self.nombre = nombre
        
        # lo mismo para el resto de parámetros
        self.apellidos = apellidos
        self.edad = edad
        self.posicion = posicion
        self.año = año
        self.vacaciones = vacaciones
        self.herramientas = herramientas
        
    def descripcion (self):
        return f"El/ella, es {self.nombre}, {self.posicion}, quien lleva con nostros en la empresa desde {self.año}. Además este año le quedan {self.vacaciones} dias de vacaciones"
    
    
    
    def cambiando_posicion(self):
        for i in self.herramientas: 
            if i.lower() == 'python':
                return f'Bien hecho {self.nombre} 🌊'
                
            elif i.lower() == "excel":
                return f"por favor {self.nombre}, deja de usar excel y aprende python con nosotros"
            
            else:
                return "no debes usar nada asi que... no decimos nada"
            
    def calcular_antiguedad(self):
        años = datetime.now().year - self.año
        meses = datetime.now().month - self.año
        return f"{self.nombre} lleva {años} años y {meses} meses en la empresa."
    

    def solicitar_aumento_vacaciones(self, dias_solicitados):
        # Lógica para procesar la solicitud de aumento de vacaciones
        if dias_solicitados > 0:
            self.vacaciones += dias_solicitados
            return f"Solicitud aprobada. Ahora tienes {self.vacaciones} días de vacaciones."
        else:
            return "Solicitud no válida. Debes solicitar un aumento positivo."
        
    def registrar_herramienta(self, nueva_herramienta):
        # Lógica para registrar una nueva herramienta en la lista
        if nueva_herramienta.lower() not in self.herramientas:
            self.herramientas.append(nueva_herramienta.lower())
            return f"¡Excelente! Ahora sabes cómo usar {nueva_herramienta}."
        else:
            return f"Ya has registrado {nueva_herramienta}. ¡Sigue aprendiendo!"
        
    def cambiar_posicion(self, nueva_posicion, nuevo_salario):
        # Lógica para cambiar la posición y salario del empleado
        self.posicion = nueva_posicion
        self.salario = nuevo_salario
        return f"¡Felicidades, {self.nombre}! Has sido promovido a {nueva_posicion} con un salario de {nuevo_salario} al año."
    

    def tomar_vacaciones(self):
        try:
            dias_solicitados = int(input("¿Cuántos días de vacaciones deseas tomar?: "))
            while dias_solicitados <= 0:
                print("Debes solicitar al menos un día de vacaciones.")
                dias_solicitados = int(input("¿Cuántos días de vacaciones deseas tomar?: "))
            
            if dias_solicitados <= self.vacaciones:
                self.vacaciones -= dias_solicitados
                return f"Disfruta tus {dias_solicitados} días de vacaciones, {self.nombre}. Te quedan {self.vacaciones} días restantes."
            else:
                return f"No tienes suficientes días de vacaciones. Te quedan {self.vacaciones} días."
        except ValueError:
            return "Error: Ingresa un número válido de días de vacaciones."

    def solicitar_aumento_salario(self):
        try:
            aumento = float(input("¿Cuánto aumento de salario deseas solicitar en porcentaje?: "))
            while aumento < 0:
                print("El aumento debe ser un valor positivo o igual a cero.")
                aumento = float(input("¿Cuánto aumento de salario deseas solicitar en porcentaje?: "))
            
            nuevo_salario = self.salario + (self.salario * aumento / 100)
            return f"Solicitud de aumento aprobada. Tu nuevo salario es de ${nuevo_salario:.2f} al año."
        except ValueError:
            return "Error: Ingresa un valor válido para el aumento de salario."

    def realizar_evaluacion(self):
        try:
            resultado = input("¿Cómo calificarías tu desempeño (Excelente/Bueno/Regular/Malo)?: ")
            resultado = resultado.lower()
            while resultado not in ["excelente", "bueno", "regular", "malo"]:
                print("Por favor, ingresa una calificación válida.")
                resultado = input("¿Cómo calificarías tu desempeño (Excelente/Bueno/Regular/Malo)?: ")
                resultado = resultado.lower()
            
            if resultado == "excelente":
                return "¡Felicidades! Has obtenido una calificación excelente. Tu desempeño ha sido destacado."
            elif resultado == "bueno":
                return "Tu desempeño ha sido calificado como bueno. Continúa así."
            elif resultado == "regular":
                return "Tu desempeño es calificado como regular. Hay espacio para mejorar."
            else:
                return "Tu desempeño ha sido calificado como malo. Debes tomar medidas para mejorar."
        except Exception:
            return "Error: Algo salió mal durante la evaluación. Inténtalo de nuevo más tarde."





            
    
    
```

```python
class Gerente(Empleados):
    def __init__(self, nombre, apellidos, edad, año, vacaciones, herramientas, departamento, empleados_a_cargo):
        # Llamamos al constructor de la clase base 'Empleados'
        super().__init__(nombre, apellidos, edad, "Gerente", año, vacaciones, herramientas)
        
        # Atributos específicos de la clase Gerente
        self.departamento = departamento
        self.empleados_a_cargo = empleados_a_cargo

    # Métodos específicos de la clase Gerente
    def asignar_tareas(self, tarea, empleado):
        return f"El Gerente {self.nombre} ha asignado la tarea '{tarea}' a {empleado}."

    def revisar_desempeno_empleado(self, empleado):
        return f"El Gerente {self.nombre} está revisando el desempeño de {empleado}."

    def gestionar_empleados(self, accion, empleado):
        if accion == "contratar":
            self.empleados_a_cargo.append(empleado)
            return f"{empleado} ha sido contratado por el Gerente {self.nombre}."
        elif accion == "despedir" and empleado in self.empleados_a_cargo:
            self.empleados_a_cargo.remove(empleado)
            return f"{empleado} ha sido despedido por el Gerente {self.nombre}."
        else:
            return f"Acción no válida para la gestión de empleados."
```
