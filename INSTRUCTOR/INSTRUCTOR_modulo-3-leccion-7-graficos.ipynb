# Graficos!

```python
# primero nuestras librerias!

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd

# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
# la informacion a trabajar

df = pd.read_csv("bank-additional-clean-nonulls.csv")
df.head(2)
```

## A-  ANALSIS DE VARIABLES NUMERICAS

```python
# 1- HISTOGRAMAS (representar la distribución de una variable numérica)

sns.histplot(data=df,
             x="age",
             color="g",
             kde=True,
             bins=20);
```

```python
# con matplot

plt.hist(x="age",
         data=df,
         bins=20,
         edgecolor="green",
         range=[20,100])


# ARMAR NUESTROS TITULOS DE GRAFICO Y DE LOS EJES

plt.xlabel("eje x de algo",fontsize = 5)
plt.ylabel("eje y de algo",fontsize = 10,loc="center")
plt.title("titulo de mi super grafico!!!");
```

```python
# 2- GRAFICOS DE DISPERSION (para identificar patrones, tendencias o correlaciones entre dos variables)

sns.scatterplot(data=df,
                x="pdays",
                y="age",
                marker ="v",
                size=10,
                color="red");

plt.xlabel("eje x PDAYS")
plt.ylabel("eje y AGE")
plt.title("naves alienigenas!!!");
```

```python
# 3- GRAFICOS DE REGRESION (cuando queramos explorar la relación entre dos variables numéricas y al mismo tiempo, ver cómo se ajusta una línea de regresión lineal a esos datos)

sns.regplot(data=df,
            x="pdays",
            y="age");
```

```python
# 4- VARIOS GRAFICOS

fig, axes = plt.subplots(nrows= 2, ncols= 2,figsize=(40,20), sharex=False, sharey=False)

sns.regplot(data=df,
            x="pdays",
            y="age",
            ax=axes[0,1])

axes[1,0].hist(x="age",
         data=df,
         bins=20,
         edgecolor="green",
         range=[20,100]);

axes[0,1].set_xlabel("nombre de eje x primer grafico",fontsize=15)


axes[1,0].set_ylabel("etiqueta de silvia",fontsize=9)

fig.suptitle("graficos en general", fontsize=20)

fig.savefig("GRAFICO_HORRIBLE.png");
```

```python
# 5- BOXPLOT (Un Boxplot de Matplotlib muestra la distribución de datos a través de sus cuartiles, resaltando medianas y posibles valores atípicos)

sns.boxplot(data=df, 
            y="age",
            color="red");
```

```python
# mismo grafico....pero con matplotlib!!!

plt.boxplot(data=df,
            x="age",
            boxprops={"color":"blue"},
            capprops={"color":"red"})
```

```python
# 6- VIOLINPLOT (para visualizar la distribución de datos numéricos en diferentes categorías o grupos)

sns.violinplot(data=df,
               y="age");
```

## B- ANALSIS DE VARIABLES CATEGORICAS

```python
df["education"].unique()
```

```python
# 1- COUNTPLOT (se utiliza para mostrar la frecuencia de observaciones en diferentes categorías)

sns.countplot(data=df,
              x = "education",
              hue = "y")


plt.xticks(rotation = 45);
```

```python
# 2- PIEPLOT (para mostrar la proporción de diferentes partes en relación con un todo)

# primero el groupby

agrupado = df.groupby("education")["id"].count().reset_index()

agrupado
```

```python
plt.pie("id",
        labels="education",
        data=agrupado,
        autopct="%1.1f%%");
```
