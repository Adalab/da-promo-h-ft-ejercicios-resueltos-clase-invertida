# Criterio de uso de media o mediana.

1) Cuándo usar la media
*	Datos simétricos: La media es adecuada cuando los datos están distribuidos simétricamente, sin sesgo significativo.
*	Sin valores atípicos (outliers): Si los datos no tienen valores atípicos extremos, la media puede ser una buena representación central.

Ejemplo: Si tienes datos de alturas de personas en una muestra donde las alturas no tienen valores extremadamente altos o bajos, la media puede ser apropiada.

2) Cuándo usar la mediana
*	Datos asimétricos: La mediana es preferible cuando los datos están sesgados o no son simétricos.
*	Con valores atípicos (outliers): La mediana es robusta frente a valores atípicos, ya que no se ve afectada por ellos. Esto la convierte en una mejor opción cuando hay outliers en los datos.

Ejemplo: Si tienes datos de ingresos donde unos pocos individuos ganan significativamente más que el resto, la mediana será una mejor representación central que la media.
