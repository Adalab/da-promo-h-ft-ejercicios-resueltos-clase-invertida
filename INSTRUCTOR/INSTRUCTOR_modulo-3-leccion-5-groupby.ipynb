```python
# quiere 3 etiquetas....pocos, algunos, y muchos kilometros
```

```python
def etiqueta_kilometros(numeros):
    if numeros < 15000:
        return "pocos kilometros"
    elif numeros > 15000 and numeros < 30000:
        return "algunos kilometros"
    elif numeros > 30000:
        return "muchos kilometros"
```

```python
df2["Kilometros"].apply(etiqueta_kilometros)
```

```python
df2["etiqueta_km"] = df2["Kilometros"].apply(etiqueta_kilometros)
```

```python
df2.head(5)
```

```python
df2["km por mil"] = df2["Kilometros"].apply(lambda x : x / 1000)
```
