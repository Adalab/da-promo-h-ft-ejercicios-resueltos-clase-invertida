# Estadistica descriptiva.

Que es la estadistica descriptiva? es la recopilacion ,resumen ,organizacion y presentacion de datos!

- Medidas de centralizacion -> responde la pregunta de donde se encuentra el valor mas tipico o el promedio de mi conjunto de datos.

        1-media: el promedio de mi conjunto de datos. la suma de todos mis valores dividido la cantidad de registros/valores (metodo .mean())

        2-mediana: es el valor que se encuentra en el medio cuando el conjunto de datos se encuentra ordenado de menor a mayor. en caso de registros pares seria la media de los dos valores centrales.

        3-moda:es el valor mas frecuente.

- Medidas de dispersion -> responde a la pregunta de que tanto estan dispersos mis datos!

        1-la varianza: valor unico que expresa la dispersion de mis datos, y su medida esta al cuadrado.

        2-la desviacion standard: medida que indica cuanto varian los datos respecto a la media. es la raiz cuadrada de la varianza,su unidad se expresa en la misma escala que el valor original de referencia.

        3-Rango: difrencia entre el valor mas alto y el mas bajo en un conjunto de datos



- Medidas de correlacion -> nos ayuda a ver la relacion entre dos variables (metodo .corr(method="COLOCAR ACA METODO SEGUN CORRESPONDA"))

        -entre 2 o mas variables y se puede hacer con continuas (una variable del tipo numerica), ordinales, categoricas.

        - entre que valores pueden tomar esas relaciones? -1 y 1 

        - relacion -1 indicaria una relacion inversa (una sube, la otra baja)

        - relacion 1 indcaria que ambas van en el mismo sentido (una sube y la otra acompaÃ±a)

        metodos para correlacion:
            1-Pearson: Se utiliza cuando la relacion entre las variables son del tipo lineal, sensible a los outliers (valor atipico).

            2-Spearman: Nos vale para relaciones lineales y no lineales (asume cualquier tipo de relacion), es menos sensible a los outliers.
            
            3-Kendall: lo utilizaremos cuando las variables tengan un "orden" (satisfaccion al cliente , de 1 a 5 (por ejemplo))


- Medidas relativas -> nos ayuda a comprender la posicion de los elemetos dentro de mi grupo de datos

        1-percentiles: es la division de mis datos en un grupo de 100 partes iguales

        2-cuartiles: divide mis datos en 4 partes iguales (q1,q2,q3,q4)
        
        3-deciles: divide mis datos en 10 partes iguales.

        extra!: rangos intercuartiles son los grupos de rangos que comprenden mi informacion 
        si tengo los datos comprendidos entre Q1 Y Q3 mi rango intercuartil es de IQR(Q1-Q3)

- Medidas categoricas

        1- Frecuencia absoluta: es el recuento de cuantas veces aparece un cierto valor dentro de mis datos.

        2- Frecuencia relativa: es el PORCENTAJE de apariciones de los elementos dentro de mis datos

        3- tabla de contingencia: es una tabla que muestra la frecuencia de ocurrencias o apariciones de un conjunto de datos elegidos por mi.

        4- moda: el valor que mas aparece!.
 

```python
#primero...las librerias!

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

## Correlaciones

```python
#veamos los tipos de correlaciones

# GeneraciÃ³n de datos
np.random.seed(42)  # Para reproducibilidad

# RelaciÃ³n positiva
x_pos = np.linspace(0, 10, 100)
y_pos = 2 * x_pos + np.random.normal(0, 2, 100)

# RelaciÃ³n negativa
x_neg = np.linspace(0, 10, 100)
y_neg = -2 * x_neg + np.random.normal(0, 2, 100)

# RelaciÃ³n nula
x_null = np.linspace(0, 10, 100)
y_null = np.random.normal(0, 2, 100)

# Crear DataFrames
df_pos = pd.DataFrame({'x': x_pos, 'y': y_pos})
df_neg = pd.DataFrame({'x': x_neg, 'y': y_neg})
df_null = pd.DataFrame({'x': x_null, 'y': y_null})

# Crear los grÃ¡ficos scatterplot con lÃ­nea de regresiÃ³n
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Scatterplot de relaciÃ³n positiva con lÃ­nea de regresiÃ³n
sns.regplot(x='x', y='y', data=df_pos, ax=axs[0], color="blue")
axs[0].set_title('RelaciÃ³n Positiva',fontsize=20)
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# Scatterplot de relaciÃ³n negativa con lÃ­nea de regresiÃ³n
sns.regplot(x='x', y='y', data=df_neg, ax=axs[1], color="red")
axs[1].set_title('RelaciÃ³n Negativa',fontsize=20)
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')

# Scatterplot de relaciÃ³n nula con lÃ­nea de regresiÃ³n
sns.regplot(x='x', y='y', data=df_null, ax=axs[2], color="green")
axs[2].set_title('RelaciÃ³n debil',fontsize=20)
axs[2].set_xlabel('X')
axs[2].set_ylabel('Y')

# Mostrar grÃ¡ficos
plt.tight_layout()
plt.show()
```

# fuerza y sentido de correlaciones.

```python
# GeneraciÃ³n de datos con las correlaciones deseadas
np.random.seed(42)

# CorrelaciÃ³n positiva de 0.85
mean_pos = [0, 0]
cov_pos = [[1, 0.85], [0.85, 1]]
x_pos, y_pos = np.random.multivariate_normal(mean_pos, cov_pos, 100).T
df_pos = pd.DataFrame({'x': x_pos, 'y': y_pos})

# CorrelaciÃ³n negativa de -0.70
mean_neg = [0, 0]
cov_neg = [[1, -0.70], [-0.70, 1]]
x_neg, y_neg = np.random.multivariate_normal(mean_neg, cov_neg, 100).T
df_neg = pd.DataFrame({'x': x_neg, 'y': y_neg})

# CorrelaciÃ³n nula de 0.15
mean_null = [0, 0]
cov_null = [[1, 0.15], [0.15, 1]]
x_null, y_null = np.random.multivariate_normal(mean_null, cov_null, 100).T
df_null = pd.DataFrame({'x': x_null, 'y': y_null})

# Calcular coeficientes de correlaciÃ³n de Pearson
corr_pos = df_pos.corr().iloc[0, 1]
corr_neg = df_neg.corr().iloc[0, 1]
corr_null = df_null.corr().iloc[0, 1]

# Crear los grÃ¡ficos scatterplot con lÃ­nea de regresiÃ³n
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Scatterplot de correlaciÃ³n positiva
sns.regplot(x='x', y='y', data=df_pos, ax=axs[0], color="blue")
axs[0].set_title(f'CorrelaciÃ³n Positiva\nCoeficiente de CorrelaciÃ³n: {corr_pos:.2f}', fontsize=20)
axs[0].set_xlabel('Variable X', fontsize=15)
axs[0].set_ylabel('Variable Y', fontsize=15)

# Scatterplot de correlaciÃ³n negativa
sns.regplot(x='x', y='y', data=df_neg, ax=axs[1], color="red")
axs[1].set_title(f'CorrelaciÃ³n Negativa\nCoeficiente de CorrelaciÃ³n: {corr_neg:.2f}', fontsize=20)
axs[1].set_xlabel('Variable X', fontsize=15)
axs[1].set_ylabel('Variable Y', fontsize=15)

# Scatterplot de correlaciÃ³n nula
sns.regplot(x='x', y='y', data=df_null, ax=axs[2], color="green")
axs[2].set_title(f'CorrelaciÃ³n Nula\nCoeficiente de CorrelaciÃ³n: {corr_null:.2f}', fontsize=20)
axs[2].set_xlabel('Variable X', fontsize=15)
axs[2].set_ylabel('Variable Y', fontsize=15)

# Ajustar tamaÃ±o de fuente de las etiquetas de los ejes
for ax in axs:
    ax.tick_params(axis='both', which='major', labelsize=15)

# Mostrar grÃ¡ficos
plt.tight_layout()
plt.show()
```

# Relacion lineal y no lineal.

```python
# GeneraciÃ³n de datos
np.random.seed(42)  # Para reproducibilidad

# RelaciÃ³n lineal
x_linear = np.linspace(0, 10, 100)
y_linear = 2 * x_linear + np.random.normal(0, 2, 100)

# RelaciÃ³n no lineal mÃ¡s pronunciada (cÃºbica)
x_nonlinear = np.linspace(0, 10, 100)
y_nonlinear = x_nonlinear**3 + np.random.normal(0, 20, 100)

# Crear DataFrames
df_linear = pd.DataFrame({'x': x_linear, 'y': y_linear})
df_nonlinear = pd.DataFrame({'x': x_nonlinear, 'y': y_nonlinear})

# Ajustar modelo cÃºbico a los datos no lineales
poly_fit = np.polyfit(df_nonlinear['x'], df_nonlinear['y'], 3)
poly_line = np.poly1d(poly_fit)(df_nonlinear['x'])

# Crear los grÃ¡ficos scatterplot con lÃ­nea de regresiÃ³n
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

# Scatterplot de relaciÃ³n lineal con lÃ­nea de regresiÃ³n
sns.regplot(x='x', y='y', data=df_linear, ax=axs[0], color="blue")
axs[0].set_title('RelaciÃ³n Lineal', fontsize=20)
axs[0].set_xlabel('Variable X', fontsize=15)
axs[0].set_ylabel('Variable Y', fontsize=15)

# Scatterplot de relaciÃ³n no lineal con ajuste cÃºbico
sns.scatterplot(x='x', y='y', data=df_nonlinear, ax=axs[1], color="red")
axs[1].plot(df_nonlinear['x'], poly_line, color="orange", label='Ajuste no lineal (cÃºbico)')
axs[1].set_title('RelaciÃ³n No Lineal', fontsize=20)
axs[1].set_xlabel('Variable X', fontsize=15)
axs[1].set_ylabel('Variable Y', fontsize=15)
axs[1].legend()

# Ajustar tamaÃ±o de fuente de las etiquetas de los ejes
for ax in axs:
    ax.tick_params(axis='both', which='major', labelsize=15)

# Mostrar grÃ¡ficos
plt.tight_layout()
plt.show()
```

## Matriz de correlacion.

```python
# Generar matriz de correlaciÃ³n aleatoria
np.random.seed(42)
correlation_matrix = np.random.rand(5, 5)

# Establecer la diagonal principal de la matriz de correlaciÃ³n a 1.0
np.fill_diagonal(correlation_matrix, 1.0)

# Establecer las relaciones negativas en ambos sentidos
correlation_matrix[1, 4] = -0.8  # "Ingresos" y "Personal"
correlation_matrix[4, 1] = -0.8  # "Personal" y "Ingresos"
correlation_matrix[0, 3] = -0.6  # "Ventas" y "ProducciÃ³n"
correlation_matrix[3, 0] = 0.6   # "ProducciÃ³n" y "Ventas"
correlation_matrix[2, 3] = -0.7  # "Gastos" y "ProducciÃ³n"

# Reflejar simÃ©tricamente la matriz de correlaciÃ³n a lo largo de la diagonal
correlation_matrix = (correlation_matrix + correlation_matrix.T) / 2

# Nombres de variables ficticias en espaÃ±ol
nombres_variables = ['Ventas', 'Ingresos', 'Gastos', 'ProducciÃ³n', 'Personal']

# Graficar la matriz de correlaciÃ³n como un mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de CorrelaciÃ³n', fontsize=20)
plt.xlabel('Variables', fontsize=15)
plt.ylabel('Variables', fontsize=15)
plt.xticks(ticks=np.arange(0.5, 5.5, 1), labels=nombres_variables)
plt.yticks(ticks=np.arange(0.5, 5.5, 1), labels=nombres_variables)
plt.tight_layout()
plt.show()
```

```python
# Generar matriz de correlaciÃ³n aleatoria
np.random.seed(42)
correlation_matrix = np.random.rand(5, 5)

# Establecer la diagonal principal de la matriz de correlaciÃ³n a 1.0
np.fill_diagonal(correlation_matrix, 1.0)

# Establecer las relaciones negativas en ambos sentidos
correlation_matrix[1, 4] = -0.8  # "Ingresos" y "Personal"
correlation_matrix[4, 1] = -0.8  # "Personal" y "Ingresos"
correlation_matrix[0, 3] = -0.6  # "Ventas" y "ProducciÃ³n"
correlation_matrix[3, 0] = 0.6   # "ProducciÃ³n" y "Ventas"
correlation_matrix[2, 3] = -0.7  # "Gastos" y "ProducciÃ³n"

# Reflejar simÃ©tricamente la matriz de correlaciÃ³n a lo largo de la diagonal
correlation_matrix = (correlation_matrix + correlation_matrix.T) / 2

# Obtener solo la mitad superior de la matriz de correlaciÃ³n
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Graficar la matriz de correlaciÃ³n como un mapa de calor con np.triu
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de CorrelaciÃ³n', fontsize=20)
plt.xlabel('Variables', fontsize=15)
plt.ylabel('Variables', fontsize=15)
plt.xticks(ticks=np.arange(0.5, 5.5, 1), labels=nombres_variables)
plt.yticks(ticks=np.arange(0.5, 5.5, 1), labels=nombres_variables)
plt.tight_layout()
plt.show()
```
